Q.1) Write a program to find all word types that occur at least 100 times in the Brown Corpus. How many of them are there?

**Importing NLTK**

````python
import nltk
````

````python
nltk.download('book', quiet=True)
````

**Accessing Brown corpus words**

````python
nltk.corpus.brown.words()
````

````python
brown_words = nltk.corpus.brown.words()
````

````python
len(brown_words)
````

````python
types = set(brown_words)
````

````python
len(types)
````

````python
freq = nltk.FreqDist(brown_words)
````

````python
freq.most_common(20)
````

**Issue:** It is considering **'The'** and **'the'** as different word types. Along with it, it is considering **'puntuation marks'** also as the types of words.

**Modifying Upper-case and Lower-case similar words to Lower-case**

````python
brown_words_lower_case = [w.lower() for w in brown_words]
````

**Removal of Punctuation symbols** - Here we are asked to find word types in the main question. But **according to nltk book**, as punctuation marks are included, these are considered as **'types'** and **not 'word types'**.

````python
brown_words_puntuations_removed = [w for w in brown_words_lower_case if w.isalpha()]
brown_words_puntuations_removed = nltk.FreqDist(brown_words_puntuations_removed)
brown_words_puntuations_removed.most_common(50)
````

````python
all_word_types = brown_words_puntuations_removed.most_common()
````

**Word types which occur atleast 100 times in the brown corpus**

````python
word_types = [(w,x) for (w,x) in all_word_types if(x>=100)]
print(word_types)
````

**Count of all word types which occur atleast 100 times in the brown corpus**

````python
len(word_types)
````

**Answer:** There are a total of **1031 word types** which occur at least 100 times in the Brown Corpus.

```
sorted(set(w.lower() for w in text1))
```
1) The above operation `w.lower()` goes through each word in text1, assigning each one in turn to the variable w and performing the specified operation of lower-case on the variable.

2) After getting final list of all words in lower-case, `'set'` function further filters the list to only unique tokens and removes all duplicated values.

3) Here, the duplicated values are removed, by use of `set` function.

4) Hence, we get the final result as 17231, which is less than that of 'Word_Types' = 19317.

````python
result_2 = sorted(w.lower() for w in set(text1))
len(result_2)
````

```
sorted(w.lower() for w in set(text1))
```
1) The above operation `set` goes through each word in text1, and filters out only the unique tokens from the entire text1. This result contains duplicated values of specific tokens (eg. 'The' and 'the').

2) Further, the `w.lower()` operation goes through each word in text1, assigning each one in turn to the variable w and performing the specified operation of lower-case on the resultant variables from `set` operation.

3) The final result has got the lower-case values of all the unique tokens in which most of the tokens are duplicated, because we performed the `set` operation earlier and then later executed the `w.lower()` operation. (eg. 'The' and 'the').

4) Hence, final result contains the same values as that of `'Word_Types'` i.e. 19317.

Q.5) Read in the texts of the State of the Union addresses, using the `state_union` corpus reader. Count occurrences of *men*, *women*, and *people* in each document. What has happened to the usage of these words over time?

**Referring - State Union Corpus**

````python
from nltk.corpus import state_union
````

**Displaying Words**

````python
nltk.corpus.state_union.words()
````

**Lower-case all words**

````python
words_state_union = [w.lower() for w in state_union.words()]
words_state_union
````

**Count of occurances of 'men' word**

````python
words_state_union.count('men')
````

**Count of occurances of 'women' word**

````python
words_state_union.count('women')
````

**Count of occurances of 'people' word**

````python
words_state_union.count('people')
````

**Details of occurances of 'men' word**

````python
men = []
for i in state_union.fileids():
  men_count = len([j for j in state_union.words(i) if(j == 'men' or j=='Men')])
  print(i,"has word men ",men_count, " times")
  men.append(men_count)
````

**Details of occurances of 'women' word**

````python
women = []
for i in state_union.fileids():
  women_count = len([j for j in state_union.words(i) if(j == 'women' or j=='Women')])
  print(i,"has word women ",women_count, " times")
  women.append(women_count)
````

**Details of occurances of 'people' word**

````python
people = []
for i in state_union.fileids():
  people_count = len([j for j in state_union.words(i) if(j == 'people' or j=='People')])
  print(i,"has word people ",people_count, " times")
  people.append(people_count)
````

````python
file_names = []
for i in state_union.fileids():
  file_names.append(i[:4])
````

**Graph : men word**

````python
import matplotlib.pyplot as plt
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(file_names, men)
plt.xlabel('Files')
plt.ylabel('Count Of word Men')
plt.title("word men over time")
plt.xticks(rotation=90)
plt.tick_params(axis='x', which='major', labelsize=10)
plt.show()
````

**Graph : women word**

````python
import matplotlib.pyplot as plt
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(file_names, women)
plt.xlabel('Files')
plt.ylabel('Count Of word women')
plt.title("word women over time")
plt.xticks(rotation=90)
plt.tick_params(axis='x', which='major', labelsize=10)
plt.show()
````

````python
import matplotlib.pyplot as plt
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(file_names, people)
plt.xlabel('Files')
plt.ylabel('Count Of word people')
plt.title("word people over time")
plt.xticks(rotation=90)
plt.tick_params(axis='x', which='major', labelsize=10)
plt.show()
````

Q.6) Write a program to guess the number of syllables contained in a text, making use of the CMU Pronouncing Dictionary.

**Referring - CMU Pronouncing Dictionary**

````python
entries = nltk.corpus.cmudict.entries()
````

**Definition of variables and letter incremental counter array**

````python
pron = []
word = []
for i in range(len(entries)):
  pron.append(entries[i][1])
  word.append(entries[i][0])
````

**Defining the function and its logic**

````python
def syllable(text):
  if text in word:
    pron_idx = word.index(text)

  stress = [char for phone in pron[pron_idx] for char in phone if char.isdigit()]
  # print(stress)
  k = 0
  for i in stress:
    # print(i)
    k1 = int(i)
    # print(k1)
    k = k + k1
    # print(k)
  return k
````

**Checking Output**

````python
syllable('firearm')
````

````python
syllable('mouse')
````
